**Exploratory Data Analysis on a Sample Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step 1: Load the Dataset**

```python
# Load the dataset (adjust the file path if necessary)
df = pd.read_csv("Titanic-Dataset - Titanic-Dataset.csv")

# Display the first few rows of the dataset to get an overview
df.head()
```

**Step 2: Basic Data Overview**

```python
print("Shape of the dataset:", df.shape)
df.info()
df.describe(include='all')
```

**Step 3: Check for Missing Values**

```python
print(df.isnull().sum())
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

**Step 4: Univariate Analysis**

```python
sns.histplot(df['Age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.show()
```

**Count of Categorical Variables (e.g., Survived, Sex):**

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
sns.countplot(x='Sex', data=df)
plt.title("Gender Count")
plt.show()
```

**Step 5: Bivariate Analysis**

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

**Survival by Passenger Class**

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

**Age Distribution by Class**

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Age Distribution by Class")
plt.show()
```

**Step 6: Correlation Analysis**

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
